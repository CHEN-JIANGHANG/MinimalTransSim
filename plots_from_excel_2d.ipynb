```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#READ DATA
file_position="data/2016126153221_passengers.csv"

with open(file_position , "r") as file :
    content=file.read()
    print("succeed")
data=content.split("\n")
legend=data[0].split(",")
del data[0]
for i,li in enumerate(legend):
    print(i," : ",li)
```

```python
#DECIDE WHAT TO PLOT
X=6
Y=12
```

```python
#PLOT

x,y=[],[]
for line in data:
    val=line.split(",")
    if len(val)>1 and val[X] is not "" and val[Y] is not "":
        x.append(float(val[X]))
        y.append(float(val[Y]))
        
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x,y)
ax.set_xlabel(legend[X])
ax.set_ylabel(legend[Y])
```
