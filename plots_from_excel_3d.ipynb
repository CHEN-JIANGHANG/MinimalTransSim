```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#READ DATA
file_position="data/example/2016166165724_general.csv"

with open(file_position , "r") as file :
    content=file.read()
    print("succeed")
data=content.split("\n")
legend=data[0].split(",")
del data[0]
for i,li in enumerate(legend):
    print(i," : ",li)
```

```python
#DECIDE WHAT TO PLOT
X=1
Y=2
Z=4
```

```python
#PLOT

x,y,z=[],[],[]
for line in data:
    val=line.split(",")
    if len(val)>1:
        x.append(float(val[X]))
        y.append(float(val[Y]))
        z.append(float(val[Z]))
        
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#ax.scatter(x,y,z)
ax.plot_trisurf(x,y,z)
ax.set_xlabel(legend[X])
ax.set_ylabel(legend[Y])
ax.set_zlabel(legend[Z])
```

```python
#SAVE THE FIGURE
name=file_position[0:-4]+"_"+legend[Z]+"_VS_"+legend[X]+"_and_"+legend[Y]
plt.savefig(name)
```
